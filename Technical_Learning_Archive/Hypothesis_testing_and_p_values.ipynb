You are interested in the relationship between `verified_status` and `video_view_count`. One approach is to examine the mean value of `video_view_count` for each group of `verified_status` in the sample data.

```python
# Compute the mean `video_view_count` for each group in `verified_status`
### YOUR CODE HERE ###
data.groupby('verified_status').agg({'video_view_count':'mean'})

```

```python
data_verified = data[data['verified_status'] == 'verified']
data_unverified = data[data['verified_status'] == 'not verified']
```

### **Task 3. Hypothesis testing**

Before you conduct your hypothesis test, consider the following questions where applicable to complete your code response:

1. Recall the difference between the null hypothesis and the alternative hypotheses. What are your hypotheses for this data project?

==> Ho: There is no difference in the average view counts between verified and unverified accounts. 
    Ha: There is a difference in the average view counts between verified and unverified accounts. 



Your goal in this step is to conduct a two-sample t-test. Recall the steps for conducting a hypothesis test:


1.   State the null hypothesis and the alternative hypothesis
2.   Choose a signficance level
3.   Find the p-value
4.   Reject or fail to reject the null hypothesis



==> ENTER YOUR NULL AND ALTERNATIVE HYPOTHESES HERE (Double Click)



You choose 5% as the significance level and proceed with a two-sample t-test.

```python
# Conduct a two-sample t-test to compare means



### YOUR CODE HERE ###
significance_level = 0.05 

stats.ttest_ind(data_verified['video_view_count'], data_unverified['video_view_count'], equal_var = True)
```

```python
# view, like, comment, share, download 
for i in ['video_view_count', 'video_like_count', 'video_share_count', 'video_comment_count', 'video_download_count']:
    tstat, pvalue = stats.ttest_ind(data_verified[i], data_unverified[i], equal_var = True)
    print(f'{i} p vlaue: {pvalue}')
    
    
```

```python
print(data_unverified.mean())
print(data_verified.mean())
```

```python
# a visualization(verified and unverified user/claim or opinion/ counts)

grouped = data_verified['claim_status'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(8, 4))  # 1 row, 2 columns


# Verified accounts
grouped = data_verified['claim_status'].value_counts()
axes[0].pie(grouped, labels = ['opinion', 'claim'],autopct='%1.1f%%',startangle = 90)
axes[0].set_title('Verified Accounts')
axes[0].axis('equal')

# Unverified accounts
grouped = data_unverified['claim_status'].value_counts()
axes[1].pie(grouped, labels = ['opinion', 'claim'],autopct='%1.1f%%',startangle = 90)
axes[1].set_title('Unverified Accounts')
axes[1].axis('equal')


plt.tight_layout()
plt.show()
```

**Question:** Based on the p-value you got above, do you reject or fail to reject the null hypothesis?


==> I reject the null hypothesis for all five variables because each p-value is significantly less than 0.05.
